# Pure Python vs. Cythonized Package

```python
x = dfx.Dual(2, 3)
y = dfx.Dual(3, 4)

i = df.Dual(2, 3)
j = df.Dual(3, 4)
```

```python
# Different number of iterations to test
iterations_list = [100, 1000, 10000, 100000]

# List of operations: Sum, Subtraction, Multiplication, & Division
operations = ['+', '-', '*', '/']
# Initialize an empty list to hold the results
data = []

# Creating a figure with 4 subplots (for four different operations)
fig, axes = plt.subplots(1, 4, figsize=(20, 6))

# Looping through each operation
for idx, op in enumerate(operations):
    cython_times = []
    python_times = []
    
    # Looping over different iteration counts
    for iterations in iterations_list:
        # Timing the Cythonized implementation
        start = time.time()
        for _ in range(iterations):
            if op == '+':
                _ = x + y
            elif op == '-':
                _ = x - y
            elif op == '*':
                _ = x * y
            elif op == '/':
                _ = x / y
        cython_times.append(time.time() - start)
        
        # Timing the pure Python implementation
        start = time.time()
        for _ in range(iterations):
            if op == '+':
                _ = i + j
            elif op == '-':
                _ = i - j
            elif op == '*':
                _ = i * j
            elif op == '/':
                _ = i / j
        python_times.append(time.time() - start)

    # Adding results to the list with ratio and percentage slowdown
    for i, iterations in enumerate(iterations_list):
        # Compute the ratio of Cython time to Python time
        ratio = cython_times[i] / python_times[i] if python_times[i] != 0 else float('nan')

        # Calculate the percentage slowdown
        if python_times[i] != 0:
            percent_slower = ((cython_times[i] - python_times[i]) / python_times[i]) * 100
        else:
            percent_slower = float('nan')  # Handle division by zero for very small times

        # Append the result with operation, iterations, Cython time, Python time, ratio, and percentage slower
        data.append([op, iterations, round(cython_times[i], 4), round(python_times[i], 4), round(ratio, 4), round(percent_slower, 2)])

    # Plotting the results for each operation
    axes[idx].plot(iterations_list, cython_times, label='Cythonized', marker='o', linestyle='-', color='b')
    axes[idx].plot(iterations_list, python_times, label='Pure Python', marker='o', linestyle='-', color='r')
    axes[idx].set_xlabel('Number of Iterations')
    axes[idx].set_ylabel('Time (seconds)')
    axes[idx].set_title(f'({op}) Operation')
    axes[idx].set_xscale('log')
    axes[idx].set_yscale('log')
    axes[idx].grid(True)
    axes[idx].legend()

# Adjusting layout to prevent overlap
plt.tight_layout()
plt.show()

# Display the results in a formatted table
header = ['Operation', 'Iterations', 'Cython Time (s)', 'Python Time (s)', 'Ratio', 'Percentage Slower']
print(f"{header[0]:<15}{header[1]:<12}{header[2]:<20}{header[3]:<20}{header[4]:<10}{header[5]:<20}")
print('-' * 100)

# Print each row of the data with four decimals and percentage slowdown
for idx, row in enumerate(data):
    print(f"{row[0]:<15}{row[1]:<12}{row[2]:<20}{row[3]:<20}{row[4]:<10.4f}{row[5]:<20.2f}")

    # Print a separator line between different operations
    if idx + 1 < len(data) and data[idx + 1][0] != row[0]:
        print('-' * 100)
```
